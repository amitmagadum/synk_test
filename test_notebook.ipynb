```python
import datetime
def _get_or_none (getter):
    try:
        obj = getter()            
        if hasattr(obj, "isDefined") and obj.isDefined():
            return obj.get()
        return obj
    except Exception:
        return None



ctx = dbutils.entry_point.getDbutils().notebook().getContext() 
host = spark.conf.get("spark.databricks.workspaceUrl") 
job_id = str(_get_or_none (ctx.jobId)).replace("(", "").replace(")", "").replace("'", "") if _get_or_none (ctx.jobId) is not None else None
job_run_id = str(_get_or_none(ctx.jobRunId)).replace("(", "").replace(")", "").replace("'", "") if _get_or_none (ctx.jobRunId) is not None else None
task_run_id = str(_get_or_none (ctx.taskRunId)).replace("(", "").replace(")", "").replace("'","") if _get_or_none (ctx.taskRunId) is not None else None
run_id = str(_get_or_none (ctx.runId)).replace("(", "").replace(")", "").replace("'","") if _get_or_none (ctx.runId) is not None else None

is_manual = not (bool (job_run_id) or bool(task_run_id)) 
user_name = spark.sql("SELECT current_user()").collect()[0][0]
notebook_path = _get_or_none(ctx.notebookPath())
workspace_host = f"https://{host}" if host and not host.startswith("https://") else host 
cluster_id = _get_or_none(ctx.clusterId)
workspace_id = spark.conf.get("spark.databricks.workspaceUrl")
#datetime = datetime.utcnow().strftime("%Y-%m-%d %H:%M") 


print(ctx, host)
print(job_id)
print(job_run_id)
print("-------")
print(task_run_id)
print(run_id)
print(is_manual)
print("-------")

print(user_name)
print(notebook_path)
print(workspace_host)
print("-------")
print(cluster_id)
print(workspace_id)
```
